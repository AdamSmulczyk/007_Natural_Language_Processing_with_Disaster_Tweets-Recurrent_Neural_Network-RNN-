 ### <a id="I.1"></a><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">I.1 Introduction</p>

**<mark style="background-color:#3b3745;color:white;border-radius:5px;opacity:1.0">Twitter</mark>** has become an important communication channel in times of emergency.
The ubiquitousness of smartphones enables people to announce an emergency they‚Äôre observing in real-time. Because of this, more agencies are interested in programatically monitoring Twitter (i.e. disaster relief organizations and news agencies).

But, it‚Äôs not always clear whether a person‚Äôs words are actually announcing a disaster. Take this example:
The author explicitly uses the word ‚ÄúABLAZE‚Äù but means it metaphorically. This is clear to a human right away, especially with the visual aid. But it‚Äôs less clear to a machine.

Disclaimer: The dataset for this competition contains text that may be considered profane, vulgar, or offensive.


Text Classification is a process involved in Sentiment Analysis. It is classification of peoples opinion or expressions into different sentiments. Sentiments include Positive, Neutral, and Negative, Review Ratings and Happy, Sad. Sentiment Analysis can be done on different consumer centered industries to analyse people's opinion on a particular product or subject.

 ### <a id="I.2"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">I.2 Dataset Description</p>

Each sample in the train and test set has the following information:

* The text of a tweet
* A keyword from that tweet (although this may be blank!)
* The location the tweet was sent from (may also be blank)

Columns
* <code>id</code> - a unique identifier for each tweet
* <code>text</code> - the text of the tweet
* <code>location</code> - the location the tweet was sent from (may be blank)
* <code>keyword</code> - a particular keyword from the tweet (may be blank)
* <code>target</code> - in train.csv only, this denotes whether a tweet is about a real disaster (1) or not (0)

 ### <a id="I.3.2"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">I.3.2 Success metric</p>

Submissions are evaluated using **<mark style="background-color:#3b3745;color:white;border-radius:5px;opacity:1.0">F1</mark>** between the predicted and expected answers.

F1 is calculated as follows:


* <code>True Positive [TP]</code> = your prediction is 1, and the ground truth is also 1 - you predicted a positive and that's true!
* <code>False Positive [FP]</code> = your prediction is 1, and the ground truth is 0 - you predicted a positive, and that's false.
* <code>False Negative [FN]</code> = your prediction is 0, and the ground truth is 1 - you predicted a negative, and that's false.

### <a id="III.1"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">III.1 Filling nulls</p>

```python
train.isnull().sum()
```

```python
# Fill missing data (for example: we can fill missing location and keyword values with "Unknown")
train['location'] = train['location'].fillna('Unknown')
train['keyword'] = train['keyword'].fillna('Unknown')
```

### <a id="III.2"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">III.2 Cleaning text</p>

```python
# Text cleaning function
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'http\S+', '', text)  # Remove links
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Apply text cleaning
train['cleaned_text'] = train['text'].apply(clean_text)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import tensorflow as tf
from tensorflow import keras
import keras_nlp
from keras.optimizers import Adam
from sklearn.metrics import f1_score
```

```python
batch_size = 32
num_training_examples = train.shape[0]
train_split = 0.8
val_split = 0.2
state=42
steps_per_epoch = int(num_training_examples)*train_split // batch_size

epochs = 2
auto = tf.data.experimental.AUTOTUNE
```

### <a id="IV.1"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">IV.1 Text processeding</p>

```python
# Stop words and stemming
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    words = text.split()
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    return ' '.join(words)

# Create a column with preprocessed text
train['processed_text'] = train['cleaned_text'].apply(preprocess_text)
# Preview the processed data
train[['text', 'cleaned_text', 'processed_text']].head()
```

### <a id="IV.2"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">IV.2 Split training and valdiation set</p>

```python
X = train["processed_text"]
y = train["target"]
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=val_split, random_state=state)
```

```python
# Load a DistilBERT model.
preset= "distil_bert_base_en_uncased"

# Use a shorter sequence length.
preprocessor = keras_nlp.models.DistilBertPreprocessor.from_preset(preset,
                                                                   sequence_length=160,
                                                                   name="preprocessor_4_tweets"
                                                                  )
# Pretrained classifier.
classifier = keras_nlp.models.DistilBertClassifier.from_preset(preset,
                                                               preprocessor = preprocessor, 
                                                               num_classes=2)
classifier.summary()
```

 ### <a id="V.1"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">V.1 Model 1</p>

```python
# Compile
classifier.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=Adam(learning_rate=1e-5),
    metrics=["accuracy"]
)

# Fit
history = classifier.fit(
    x=X_train,
    y=y_train,
    batch_size=batch_size, 
    epochs=epochs, 
    validation_data=(X_val, y_val)
)
```

 ### <a id="V.1.1"><p style="background-color: #636B2F;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">V.1.1 Model 1 - Evaluation</p>

```python
classifier.evaluate(X_val, y_val)
```

```python
train_true = y_train.copy()
train_pred = classifier.predict(X_train)
train_pred = np.argmax(train_pred, axis=1) 

val_true = y_val.copy()
val_pred = classifier.predict(X_val)
val_pred = np.argmax(val_pred, axis=1) 

train_f1_score = f1_score(train_true, train_pred)
val_f1_score = f1_score(val_true, val_pred)

print("F1-score metric on train data == {}".format(train_f1_score))
print("F1-score metric on validation dataset == {}".format(val_f1_score))
```
